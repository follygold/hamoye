# Data Cleaning

```python
#creating header for lists
lists_header = lists.iloc[0] 

lists = lists[1:] 

lists.columns = lists_header
lists
```

```python
#creating header for transactions
transactions_header = transactions.iloc[0] 

transactions = transactions[1:] 

transactions.columns = transactions_header
transactions
```

```python
#creating header for address
address_header = address.iloc[0] 

address = address[1:] 

address.columns = address_header
address
```
